```python
import pandas as pd
```

```python
# Fazendo a leitura
df = pd.read_csv('Players.csv')
df
```

```python
# ISNULL
df.isnull().sum()
```

```python
# FILLNA
df['height'].fillna(df['height'].mean(), inplace=True)
df.isnull().sum()
```

```python
# FILLNA
df['weight'].fillna(df['weight'].median(), inplace=True)
df.isnull().sum()
```

```python
# FILLNA
df['birth_state'].fillna('Indiana', inplace=True)
df.isnull().sum()
```

```python
# LOC
df.loc[(df['height'] > 200) & (df.weight > 100) & (df.birth_state == 'New York')]
```

```python
# OU
df.loc[(df['birth_state'] == 'Indiana') | (df.birth_state == 'New York') | (df.birth_state == 'Oregon')]
```

```python
# ISIN 
df.loc[df['birth_state'].isin(['Indiana', 'New York', 'Oregon'])]
```

```python
df.loc[df['birth_city'].isnull()]
```
