```python
lista = [1, 2, 3, 4, 5]

serie = pd.Series(lista)

# serie
print('Ndim',serie.ndim)
print('Shape', serie.shape)
print('Size', serie.size)
print('Média', serie.mean())
print('Mediana', serie.median())
print('Máximo', serie.max())
print('Mínimo', serie.min())
```

```python
dicio = {
    'coluna1': list(range(10)),
    'coluna2': list(range(10, 101, 10)),
    'coluna3': list(range(20, 30, 1))
}

dataframe = pd.DataFrame(dicio)
display(dataframe)
print('Ndim',dataframe.ndim)
print('Shape', dataframe.shape)
print('Size', dataframe.size)
print('Contar Valores', dataframe.value_counts())
print('Média', dataframe.mean())
print('Mediana', dataframe.median())
print('Máximo', dataframe.max())
print('Mínimo', dataframe.min())
```
